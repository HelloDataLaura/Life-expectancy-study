```python
# Creation of a class to clean data
class DataCleaner:
    def __init__(self, df):
        self.df_life = df_life

    # Drop irrelevant columns for this analysis
    def drop_columns(self, columns_to_drop):
        self.df_life = self.df_life.drop(columns=columns_to_drop)

    # Rename remaining columns for this analysis
    def rename_columns(self, new_column_names):
        self.df_life = self.df_life.rename(columns=new_column_names)

    # Remove a value from a column
    def remove_value_from_column(self, column, value_to_remove):
        self.df_life = self.df_life[self.df_life[column] != value_to_remove]

    # Change format from 'capitalized' to 'title'
    def capitalize_to_title(self, column):
        self.df_life[column] = self.df_life[column].str.title()

    def get_cleaned_df_life(self):
        return self.df_life
```

```python
# Process of cleaning data

cleaner = DataCleaner(df_life)

columns_to_remove = ['id', 'Sexo', 'Fecha creacion', 'Fecha baja']
cleaner.drop_columns(columns_to_remove)

cleaner.remove_value_from_column('Distrito', 'Todos')

cleaner.rename_columns({
    'Año': 'date_year',
    'Distrito': 'district',
    'Valor': 'age_expectancy',
    })

cleaner.capitalize_to_title('district')

# getting the cleaning dataset
df_life_cleaned = cleaner.get_cleaned_df_life()
df_life_cleaned.head(10)
```

```python
# Save clean data to a new CSV file
df_life_cleaned.to_csv('life_expc_vlc_cleaned.csv', sep=',', index=False)
```

```python
# dropping years from 2009 to 2014 as in rent dataset we only have data from 2015 to 2022
years_to_drop = [2009, 2010, 2011, 2012, 2013, 2014]
df_life_cleaned = df_life_cleaned[~df_life_cleaned['date_year'].isin(years_to_drop)]
```
